### Un modelo teórico discreto:

##### Ensayo de Bernoulli

Lanzar una moneda tiene dos resultados posibles. Este experimento simple, llamado ensayo de Bernoulli, se modela utilizando una variable aleatoria llamada Bernoulli $X$ con parámetro $p$, donde: $0\leq p\leq1$, y $X: 0,1$

$$
P(X = x) = p^x(1-p)^{1-x}
$$

#### Distribución Binomial
El número de éxitos en $n$ ensayos de Bernoulli con una probabilidad de éxito $p$ se denomina variable aleatoria binomial $X$ con parámetros $n$ y $p$, donde: $X: 0,1,...,n$
$$
P(X = x) = {n \choose x} p^x(1-p)^{n-x} \\
$$
1. Cree una función en Python con nombre "binomial" que dado los valores de $n$ y $p$, retorne las probabilidades para cada valor de la variable $X$. Utilice la distribución Binomial para hallar estas probabilidades

````python
def binomial(n,p):
    probabilidades = []
    for x in range(0,n+1):
        prob = combinat(n,x)*p**x*(1-p)**(n-x)
        probabilidades.append(prob)
    return probabilidades

def combinat(n,x):
    result = facto(n)/(facto(x)*facto(n-x))    
    return result

def facto(y):
    facto = 1
    for i in range(1,y+1):
        facto = facto*i
    return facto
 
#print(facto(5))
#print(combinat(8,2))
print (binomial(8,0.25))
````
